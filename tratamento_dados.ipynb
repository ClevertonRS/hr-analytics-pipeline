```python
cols_remover = [col for col in df.columns if df[col].nunique() == 1]
df = df.drop(columns=cols_remover)
print(f"Colunas com valores unicos removidas {cols_remover}")
```

```python
df.shape
```

```python
colunas_str = df.select_dtypes(include=['str']).columns.tolist()

print(colunas_str)
```

```python
df['Status'] = df['Attrition'].map({
    'Yes': 'Desligado',
    'No': 'Ativo'
})
```

```python
df['Department'].unique()
```

```python
df['Departamento'] = df['Department'].map({
    'Sales': 'Vendas',
    'Research & Development': 'P&D',
    'Human Resources': 'RH'
})
```

```python
df['HorasExtras'] = df['OverTime'].map({
    'Yes': 'Sim',
    'No': 'Não'
})
```

```python
df['nivelCargo'] = df['JobLevel'].map({
    1: 'Jr',
    2: 'Pleno',
    3: 'Senior',
    4: 'Especialista',
    5: 'Diretor',
})
```

```python
df = df.rename(columns={'nivelCargo': 'NivelCargo'})
```

```python
df['FaixaSalarial'] = pd.cut(
    df['MonthlyIncome'],
    bins = [0, 3000, 6000, 10000, 20000],
    labels=['Até 3k', '3k-6k', '6k-10k', 'Acima 10k'],
    ordered=True
)
```

```python
df['FaixaIdade'] = pd.cut(
    df['Age'],
    bins=[17, 25, 35, 45, 100],
    labels=['18–25', '26–35', '36–45', '46+'],
    ordered=True
)
```

```python
df['FaixaTempoEmpresa'] = pd.cut(
    df['YearsAtCompany'],
    bins=[-1, 2, 5, 10, 30],
    labels=['0-2 anos', '3-5 anos', '6-10 anos', '10+ anos'],
    ordered=True
)
```

```python
df['CustoTurnover'] = np.where(
    df['Status'] == 'Desligado',
    df['MonthlyIncome'] * 12 * 1.5,
    0
)
```

```python
tunover = df['Status'].value_counts()
tunover
```

```python
percentual = tunover['Desligado'] / len(df) * 100
print(f"Taxa de turnover: {percentual:.1f}%")
```

```python
df.groupby('Departamento')['Status'].apply(lambda x: (x == 'Desligado').sum() / len(x) * 100).round(1)
```

```python
df.groupby('HorasExtras')['Status'].apply(lambda x: (x == 'Desligado').sum() / len(x) * 100).round(1)
```
